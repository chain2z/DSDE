```python
def readCSV(filepath):
    df = pd.read_csv(filepath)
    df.index = df["Unnamed: 0"].values
    df.drop(labels= "Unnamed: 0",axis = 1,inplace=True)
    return df
```

```python
df = readCSV(f"{DATA_PATH}\\Data4Visualization.csv")
df.columns
```
