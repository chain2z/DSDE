```python
import pandas as pd
import numpy as np
import os
```

```python
PATH = os.path.dirname(os.getcwd())
DATA_PATH = f"{PATH}\\Datas"
MODEL_PATH = f"{PATH}\\Model"
PATH
```

```python
def readCSV(filepath):
    df = pd.read_csv(filepath)
    df.index = df["Unnamed: 0"].values
    df.drop(labels= "Unnamed: 0",axis = 1,inplace=True)
    return df
```

```python
df = readCSV(f"{DATA_PATH}\\forEDA.csv")
df
```

```python
", ".join(df.columns.tolist())
```

```python
nanNum = pd.DataFrame(df.isna().sum()).T
nonNanNum = pd.DataFrame((~df.isna()).sum()).T
nonNanNum.index = [1]
noneNumData = pd.concat([nanNum , nonNanNum] , axis = 0 , ignore_index=False)
```

```python
noneNumData.to_csv("missing.csv")
```

```python
citationSerie = df["citedby-count"]
citationDict = {
    "max":citationSerie.max(),
    "min":citationSerie.min(),
    "mean":citationSerie.mean(),
    "median":citationSerie.median(),
    "most-frequent":citationSerie.mode(),
    "SD":citationSerie.std(),
    "missing":citationSerie.isna().sum(),
    "present":(~citationSerie.isna()).sum(),
    "outlier":((citationSerie < citationSerie.quantile(.25)-1.5*citationSerie.std()) | (citationSerie > citationSerie.quantile(.75)+1.5*citationSerie.std())).sum()
}
citationDF = pd.DataFrame(citationDict)
citationDF
```

```python
citationDF.to_csv(f"{DATA_PATH}\\citationEDA.csv")
```

```python
byYear = df[["citedby-count" , "year"]]
year_citation_relationship = byYear.groupby("year").mean()
```
